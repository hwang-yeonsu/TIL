```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

# Normal Distribution

```python
norm_data = np.random.normal(0,1,50000)
_ = plt.hist(norm_data, bins=1000, histtype='stepfilled', density=True)
```

```python
## QQ-Plot(정규분포와 비교)
plt.figure(figsize=(10,5))
_ = stats.probplot(norm_data, dist=stats.norm, plot=plt)
plt.show()
```

# Uniform Distribution

```python
uniform_data = np.random.uniform(low=0, high=2, size=50000)
_ = plt.hist(uniform_data, bins=1000, histtype='stepfilled', density=True)
```

```python
## QQ-Plot(정규분포와 비교)
plt.figure(figsize=(10,5))
_ = stats.probplot(uniform_data, dist=stats.norm, plot=plt)
plt.show()
```

# Fat Tail
자연에서 발생하는 현상 중 많은 것들을 정규 분포를 따르는 확률 변수로 모형화하여 사용하고 있다. 그러나 이러한 데이터들을 자세히 살펴보면 실제로는 정규분포보다 양 끝단의 비중이 더 큰것을 알 수 있다. 이를 fat tail 현상이라고 한다.

예를 들어, 주식의 수익률은 보통 정규 분포를 따르는 것으로 모형화하여 사용하나 fat tail 현상을 보인다.

```python
import FinanceDataReader as fdr

data = fdr.DataReader('GOOG', '2010').Close
data = data.dropna()
data.head()
len(data)
```

```python
## log returns
goog = data.pct_change() + 1
goog = np.log(goog[1:])
goog.plot()
```

```python
_ = plt.hist(goog, bins=1000, histtype='stepfilled', density=True)

## QQ-Plot(정규분포와 비교)
plt.figure(figsize=(10,5))
_ = stats.probplot(goog, dist=stats.norm, plot=plt)
plt.show()
```
