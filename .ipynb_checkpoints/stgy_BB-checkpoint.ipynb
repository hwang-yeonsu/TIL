```python
## lower
tmp = 0
entry_flag = 0
for i in goog_BB.index:
    if(tmp > goog_BB.loc[i].lower_flag and entry_flag==0):
        trade_list.append((i, goog_BB.loc[i,'Close'], 'lower_entry'))
        entry_flag = 1
    if(entry_flag): goog_BB.loc[i, 'trade'] = '*'
    if(entry_flag and goog_BB.loc[i].Close >= goog_BB.loc[i].middle):
        trade_list.append((i, goog_BB.loc[i,'Close'], 'lower_exit'))
        entry_flag = 0
    tmp = goog_BB.loc[i].lower_flag
```

# func

```python
import pandas as pd
import numpy as np
import talib as ta
%matplotlib inline

def stgy_BB(data:'Series', fee=0.003, sl:'positive float'=0):
    upper, middle, lower = ta.BBANDS(data, timeperiod=20)
    df_BB = pd.DataFrame({'Close': data, 'upper': upper, 'middle': middle, 'lower': lower})
    
    ## set upper,lower flag
    df_BB['upper_flag'] = 0
    df_BB['lower_flag'] = 0
    upper_signal = df_BB[df_BB.Close >= df_BB.upper].index
    lower_signal = df_BB[df_BB.Close <= df_BB.lower].index
    df_BB.loc[upper_signal, 'upper_flag'] = 1
    df_BB.loc[lower_signal, 'lower_flag'] = 1
    
    trade_list=[]
    tmp = [0,0]
    entry_flag = 0
    for i in df_BB.index:
        ## entry
        if(tmp[0] > df_BB.loc[i].upper_flag and entry_flag==0):
            trade_list.append((i, df_BB.loc[i,'Close'], 'upper_entry'))
            entry_flag = -1
        elif(tmp[1] > df_BB.loc[i].lower_flag and entry_flag==0):
            trade_list.append((i, df_BB.loc[i,'Close'], 'lower_entry'))
            entry_flag = 1
        ## exit
        if(entry_flag==-1 and df_BB.loc[i].Close <= df_BB.loc[i].middle):
            trade_list.append((i, df_BB.loc[i,'Close'], 'upper_exit'))
            entry_flag = 0
        elif(entry_flag==1 and df_BB.loc[i].Close >= df_BB.loc[i].middle):
            trade_list.append((i, df_BB.loc[i,'Close'], 'lower_exit'))
            entry_flag = 0
        tmp = [df_BB.loc[i].upper_flag, df_BB.loc[i].lower_flag]

    ## returns & check margin call
    returns = []
    sl = float('inf') if(sl==0) else sl
    margin_calls = 0
    margin_ratios = []
    for i in range(1, len(trade_list), 2):
        open_order = df_BB.loc[trade_list[i-1][0]:trade_list[i][0]].Close
        if(trade_list[i][2]=='upper_exit'):
            sl_price = open_order[0]*(1+sl)
            if(open_order.max() >= sl_price): 
                returns.append((1-sl))
                max_margin_ratio = -sl
            else: 
                returns.append(open_order[0] / trade_list[i][1] - fee)  ## short return
                max_margin_ratio = 1 - open_order.max()/open_order[0]
            margin_ratios.append(max_margin_ratio)
        else:
            sl_price = open_order[0]*(1-sl)
            if(open_order.min() <= sl_price): 
                returns.append((1-sl))
                max_margin_ratio = -sl
            else: 
                returns.append(trade_list[i][1] / open_order[0] - fee)  ## long return
                max_margin_ratio = open_order.min()/open_order[0] - 1
            margin_ratios.append(max_margin_ratio)
        ## check margin call
        if(max_margin_ratio <= -0.8):
            margin_calls += 1
            print(f'## margin call!! - {trade_list[i-1]} ##')
    
    ## outputs
    ## margin_ratios: 각 거래의 최대 마진 비율
    print(f'\nMax Margin Ratio\n'
          f'(in open trade) {round(min(margin_ratios),3):8}\n'
          f'Liquidations    {margin_calls:8}')
    years = round((data.index[-1] - data.index[0]).days/252, 4)
    output = show_output(returns, years)
    return {'tot_return':output['tot_return'], 'mdd': output['mdd'], 'returns': returns, 
            'trade': trade_list, 'margin_ratios': margin_ratios, 'dataframe': df_BB}



def show_output(returns:list, years:int):
    r = np.array(returns)
    r_cumprod = np.cumprod(r)
    ## total return
    tot_return = r_cumprod[-1]
    ## CAGR
    cagr = tot_return**(1/years)
    ## trades
    trades = len(r)
    ## Avg. Trade
    r_avg = r.mean()
    ## win rate
    count = (r>1).sum()
    win_rate = count/trades
    ## best trade
    best_trade = max(r)
    ## worst trade
    worst_trade = min(r)
    ## MDD
    peak, drawdown = 0, []
    for i in r_cumprod:
        if(i>peak): peak = i
        else: drawdown.append(i/peak-1)
    mdd = min(drawdown)
    ## MDD Duration
    mdd_duration = (drawdown==mdd).sum()
    ## Sharpe Ratio, Profit Factor
    
    print(f'------------------------\n'
          f'Total Return    {round(tot_return,3):8}\n'
          f'CAGR[/252]      {round(cagr, 3):8}\n'
          f'Trades          {trades:8}\n'
          f'Return Avg      {round(r_avg, 3):8}\n'
          f'Win Rate        {round(win_rate*100,3):8}\n'
          f'Best Trade      {round(best_trade,3):8}\n'
          f'Worst Trade     {round(worst_trade,3):8}\n'
          f'MDD[%]          {round(mdd*100,3):8}\n'
         )
    pd.Series(r_cumprod).plot(title='Return')
    return {'tot_return':tot_return, 'mdd': mdd}


## plot BBAND with close
def BB_plot(data:'Series', start:str=None, end:str=None):
    u, m, l = ta.BBANDS(data, timeperiod=20)
    df_data = pd.DataFrame({'C': data, 'U': u, 'M': m, 'L': l})
    df_data.loc[start:end].plot()
    
    

## plot returns and hist
def return_plot(returns:list):
    tmp = pd.DataFrame(returns, columns=['returns'])
    tmp.cumprod().plot()
    tmp.plot()
    tmp.plot(kind='hist', bins=30)
```

### sort by MDD
