## plt.scatter 를 활용한 산점도

plt.scatter와 plt.plot의 주요 차이점은 scatter의 경우 각 점의 속성(크기, 표면색상, 테두리색상 등)을 개별적으로 제어하거나 데이터에 매핑할 수 있는 산점도를 만드는데 사용할 수 있다는 것

```python
plt.scatter(t, np.sin(t), marker='o');
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)

plt.scatter(x,y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  ## 색상 척도 표시
```

**이처럼 여러 색상과 특징을 나타내는 산점도는 데이터의 탐색과 표현이라는 두 가지 측면에서 모두 유용하다. 하지만 데이터셋이 수천 개가 넘어가는 큰 경우에는 plot이 scatter보다 확실히 더 효율적이다. 그 이유는 scatter는 각 점에 대한 다양한 크기와 색상을 나타내는 능력이 있어 랜더러가 각 점을 개별적으로 구성하는 추가 작업을 해야 하기 때문이다. 반면, plot에서는 점이 기본적으로 항상 서로 복제되므로 점의 모양을 결정하는 작업이 전체 데이터 집합에 대해 한 번만 수행된다.**

# 정규분포
np.random.randn - 표준정규분포(Standard normal distribution)로부터 샘플링된 난수를 반환

표준정규분포 N(1, 0)이 아닌, 평균 μ, 표준편차 σ 를 갖는 정규분포 N(μ, σ)의 난수를 생성하기 위해서는
σ * np.random.randn(…) + μ 와 같은 형태로 사용

```python
a = np.random.randn(100000)
b = 2 * np.random.randn(100000) - 1
c = 4 * np.random.randn(100000) + 2

plt.hist(a, bins=100, density=True, alpha=0.5, label='(mean, stddev)=(0, 1)')
plt.hist(b, bins=100, density=True, alpha=0.5, label='(mean, stddev)=(-1, 2)')
plt.hist(c, bins=100, density=True, alpha=0.5, label='(mean, stddev)=(2, 4)')
## density=True - 빈도가 아닌 비율 표시

plt.xlim(-15, 25)
plt.legend();
```

```python
plt.boxplot(a);
```

# 밀도 플롯과 등고선 플롯

## 3차원 함수 시각화

```python
def func(x,y):
    return np.sin(x)**10 + np.cos(10+y*x) * np.cos(x)

x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X, Y = np.meshgrid(x,y)  ## 1차원 배열로부터 2차원 그리드를 생성
Z = func(X,Y)

plt.contour(X,Y,Z, colors='black');
## 단색이 사용되면 음수 값은 점선으로, 양수 값은 실선으로 표시
```

```python
plt.contour(X,Y,Z, 20, cmap='RdGy'); ## Red-Grey
```

```python
plt.contourf(X,Y,Z, 20, cmap='RdGy')
plt.colorbar();
```

# Seaborn을 활용한 시각화

Matplotlib의 단점을 보완해 일반적인 통계 플롯 유형을 위한 간단한 고차원 함수를 정의하며, Pandas DataFrame이 제공하는 기능과 통합하는 고수준의 API를 Matplotlib 위에 제공

Seaborn을 활용하면 Matplotlib의 원시 명령어를 사용해도 내부적으로 Seaborn의 편리한 API가 수행됨

```python
## 히스토그램
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x','y'])
for col in 'xy':
    _ = plt.hist(data[col], alpha=0.5)
```

```python
## seaborn의 커널 밀도 추정 함수
for col in 'xy':
    _ = sns.kdeplot(data[col], shade=True)
```

```python
## 히스토그램과 KDE를 결합
sns.histplot(data['x'], kde=True, stat="density", linewidth=0)
sns.histplot(data['y'], kde=True, stat="density", linewidth=0, color='g');
```

```python
## 2차원 kdeplot
sns.kdeplot(data.x, y=data.y, shade=True);
```

```python
## jointplot hex
sns.jointplot(x=data.x, y=data.y, kind='hex');
```

```python
## jointplot reg
sns.jointplot(data=data, x='x', y='y', kind='reg');
```
